**Common Student Questions:**
- **Q: Which chart should I use?**
  - Answer: Histogram for distributions, scatter for relationships, box plot for outliers
  - Example: Age distribution â†’ histogram, Age vs Income â†’ scatter plot
  - Rule: Distribution â†’ histogram/box, Relationship â†’ scatter/heatmap
  
- **Q: Why visualize if I have statistics?**
  - Answer: Visuals show patterns, outliers, and relationships that statistics miss
  - Example: Statistics show mean=50, but histogram shows two peaks (bimodal)
  - Benefit: Visuals reveal insights statistics can't capture

```python
# PART 1: CREATE SAMPLE DATA | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©
# ============================================================================
print("\nğŸ“Š PART 1: Create Sample Data | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©")
print("-" * 70)

np.random.seed(42)
n = 200
df = pd.DataFrame({
    'age': np.random.normal(35, 10, n),
    'income': np.random.normal(50000, 15000, n),
    'satisfaction': np.random.uniform(1, 10, n),
    'category': np.random.choice(['A', 'B', 'C'], n)
})

# Create relationship: income correlates with age
df['income'] = df['income'] + df['age'] * 1000 + np.random.normal(0, 5000, n)

print("Sample dataset created:")
print(df.head())
print(f"\nDataset shape: {df.shape}")

# ============================================================================
```

```python
# PART 2: DISTRIBUTION VISUALIZATIONS | ØªØµÙˆØ±Ø§Øª Ø§Ù„ØªÙˆØ²ÙŠØ¹
# ============================================================================
print("\n" + "=" * 70)
print("PART 2: Distribution Visualizations | ØªØµÙˆØ±Ø§Øª Ø§Ù„ØªÙˆØ²ÙŠØ¹")
print("=" * 70)

# Histogram
print("\nâœ… Example 1: Histogram (Frequency Distribution)")
print("-" * 70)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['age'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (Histogram)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(df['income'], bins=20, edgecolor='black', alpha=0.7, color='orange')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution (Histogram)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("ğŸ’¡ Histograms show: How data is distributed, peaks, spread")

# Box Plot
print("\nâœ… Example 2: Box Plot (Quartiles and Outliers)")
print("-" * 70)
plt.figure(figsize=(8, 5))
df[['age', 'income']].boxplot()
plt.ylabel('Value')
plt.title('Box Plot: Age and Income')
plt.grid(True, alpha=0.3)
plt.show()
print("ğŸ’¡ Box plots show: Median, quartiles, outliers, spread")

# ============================================================================
```

```python
# PART 3: RELATIONSHIP VISUALIZATIONS | ØªØµÙˆØ±Ø§Øª Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 3: Relationship Visualizations | ØªØµÙˆØ±Ø§Øª Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª")
print("=" * 70)

# Scatter Plot
print("\nâœ… Example 3: Scatter Plot (Two Variables)")
print("-" * 70)
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['income'], alpha=0.6, s=50)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income (Scatter Plot)')
plt.grid(True, alpha=0.3)
plt.show()
print("ğŸ’¡ Scatter plots show: Relationships, correlations, clusters")

# Correlation Heatmap
print("\nâœ… Example 4: Correlation Heatmap (Multiple Variables)")
print("-" * 70)
correlation = df[['age', 'income', 'satisfaction']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
print("ğŸ’¡ Heatmaps show: Correlations between all variable pairs")

# ============================================================================
```

```python
# PART 4: CATEGORICAL VISUALIZATIONS | ØªØµÙˆØ±Ø§Øª Ø§Ù„ÙØ¦Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 4: Categorical Visualizations | ØªØµÙˆØ±Ø§Øª Ø§Ù„ÙØ¦Ø§Øª")
print("=" * 70)

# Bar Plot
print("\nâœ… Example 5: Bar Plot (Categorical Data)")
print("-" * 70)
category_counts = df['category'].value_counts()
plt.figure(figsize=(8, 5))
category_counts.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution (Bar Plot)')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
print("ğŸ’¡ Bar plots show: Counts, frequencies, comparisons")

# ============================================================================
# SUMMARY | Ø§Ù„Ù…Ù„Ø®Øµ
# ============================================================================
print("\n" + "=" * 70)
print("Summary | Ø§Ù„Ù…Ù„Ø®Øµ")
print("=" * 70)
print("""
âœ… What you learned:
   1. Distribution Visualizations: Histograms, box plots (understand data spread)
   2. Relationship Visualizations: Scatter plots, heatmaps (find correlations)
   3. Categorical Visualizations: Bar plots (compare categories)
   4. EDA Goals: Understand structure, find patterns, detect outliers

ğŸ¯ Key Takeaways:
   - Histograms: Show frequency distributions
   - Box plots: Show quartiles and outliers
   - Scatter plots: Show relationships between variables
   - Heatmaps: Show correlations between all variables
   - Bar plots: Show categorical frequencies

ğŸ“š Next Steps:
   - Example 6: Statistical EDA (complements visual EDA)
   - Unit 3: Data Visualization (more advanced techniques)
   - Use EDA insights for feature engineering
""")
print("âœ… EDA visualization concepts understood!")
```
