**Common Student Questions:**
- **Q: What's the difference between loc and iloc?**
  - Answer: loc uses labels, iloc uses integer positions
  - Example: df.loc[0] = row with index 0, df.iloc[0] = first row
  - Tip: Use loc for labels, iloc for positions
  
- **Q: When do I use groupby?**
  - Answer: When you need to aggregate by categories
  - Example: Group by 'category', then calculate mean
  - Benefit: Powerful for summarizing data

```python
# PART 1: CREATE SAMPLE DATA | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©
# ============================================================================
print("\nðŸ“Š PART 1: Create Sample Data | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©")
print("-" * 70)

# Create sample DataFrame
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['Ahmed', 'Sara', 'Mohammed', 'Fatima', 'Ali', 'Layla'],
    'age': [25, 30, 35, 28, 32, 27],
    'salary': [5000, 7000, 9000, 6000, 8000, 5500],
    'department': ['IT', 'HR', 'IT', 'Finance', 'IT', 'HR'],
    'experience': [2, 5, 8, 3, 6, 2]
})

print("Sample DataFrame:")
print(df)
print(f"\nDataFrame shape: {df.shape}")
print(f"DataFrame info:")
print(df.info())

# ============================================================================
```

```python
# PART 2: DATA SELECTION | Ø§Ø®ØªÙŠØ§Ø± Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 2: Data Selection | Ø§Ø®ØªÙŠØ§Ø± Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª")
print("=" * 70)

# Using loc (label-based)
print("\nâœ… Example 1: Using loc (Label-based Selection)")
print("-" * 70)
print("Select rows by label:")
print(df.loc[0:2])  # Rows 0, 1, 2
print("\nSelect specific columns:")
print(df.loc[:, ['name', 'age', 'salary']])
print("\nSelect rows and columns:")
print(df.loc[0:2, ['name', 'salary']])

# Using iloc (integer position-based)
print("\nâœ… Example 2: Using iloc (Position-based Selection)")
print("-" * 70)
print("Select first 3 rows:")
print(df.iloc[0:3])
print("\nSelect specific columns by position:")
print(df.iloc[:, [1, 2, 3]])  # Columns 1, 2, 3

# ============================================================================
```

```python
# PART 3: DATA FILTERING | ØªØµÙÙŠØ© Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 3: Data Filtering | ØªØµÙÙŠØ© Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª")
print("=" * 70)

# Boolean indexing
print("\nâœ… Example 3: Boolean Indexing (Filtering)")
print("-" * 70)
it_employees = df[df['department'] == 'IT']
print("IT Department employees:")
print(it_employees)

high_salary = df[df['salary'] > 6000]
print("\nEmployees with salary > 6000:")
print(high_salary)

# Multiple conditions
print("\nâœ… Example 4: Multiple Conditions")
print("-" * 70)
it_high_salary = df[(df['department'] == 'IT') & (df['salary'] > 6000)]
print("IT employees with salary > 6000:")
print(it_high_salary)

# ============================================================================
```

```python
# PART 4: DATA MANIPULATION | Ù…Ø¹Ø§Ù„Ø¬Ø© Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 4: Data Manipulation | Ù…Ø¹Ø§Ù„Ø¬Ø© Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª")
print("=" * 70)

# Adding columns
print("\nâœ… Example 5: Adding Columns")
print("-" * 70)
df['bonus'] = df['salary'] * 0.1  # 10% bonus
df['total_compensation'] = df['salary'] + df['bonus']
print("Added 'bonus' and 'total_compensation' columns:")
print(df[['name', 'salary', 'bonus', 'total_compensation']])

# Grouping and aggregation
print("\nâœ… Example 6: Grouping and Aggregation")
print("-" * 70)
dept_stats = df.groupby('department').agg({
    'salary': ['mean', 'sum', 'count'],
    'age': 'mean',
    'experience': 'mean'
}).round(2)
print("Statistics by department:")
print(dept_stats)

# ============================================================================
```

```python
# PART 5: DATA TRANSFORMATION | ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 5: Data Transformation | ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª")
print("=" * 70)

# Sorting
print("\nâœ… Example 7: Sorting Data")
print("-" * 70)
sorted_by_salary = df.sort_values('salary', ascending=False)
print("Sorted by salary (descending):")
print(sorted_by_salary[['name', 'salary']])

# String operations
print("\nâœ… Example 8: String Operations")
print("-" * 70)
df['name_upper'] = df['name'].str.upper()
df['name_length'] = df['name'].str.len()
print("String operations:")
print(df[['name', 'name_upper', 'name_length']])

# ============================================================================
# SUMMARY | Ø§Ù„Ù…Ù„Ø®Øµ
# ============================================================================
print("\n" + "=" * 70)
print("Summary | Ø§Ù„Ù…Ù„Ø®Øµ")
print("=" * 70)
print("""
âœ… What you learned:
   1. Data Selection: loc (labels), iloc (positions)
   2. Data Filtering: Boolean indexing, multiple conditions
   3. Data Manipulation: Add columns, group, aggregate
   4. Data Transformation: Sort, string operations

ðŸŽ¯ Key Takeaways:
   - loc: Label-based selection (use index/column names)
   - iloc: Position-based selection (use integer positions)
   - Boolean indexing: Filter rows based on conditions
   - Groupby: Aggregate data by categories
   - Pandas: Powerful tool for all data manipulation

ðŸ“š Next Steps:
   - Example 2: Data Preparation for ML (apply these skills)
   - Example 3: Implementing ML Models (needs manipulated data)
   - Practice: Manipulate real datasets
""")
print("âœ… Pandas concepts understood!")
```
