## Introduction

**Matplotlib and Seaborn** provide comprehensive tools for creating various chart types. Understanding when to use each chart type is essential for effective data visualization and communication.

```python
import matplotlib.pyplot as pltimport seaborn as snsimport pandas as pdimport numpy as np# Set styleplt.style.use('seaborn-v0_8')sns.set_palette("husl")print("âœ… Libraries imported!")print("\n" + "=" * 70)print("Chart Types: Matplotlib and Seaborn | Ø£Ù†ÙˆØ§Ø¹ Ø§Ù„Ù…Ø®Ø·Ø·Ø§Øª")print("=" * 70)# ============================================================================# PART 1: CREATE SAMPLE DATA | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©# ============================================================================print("\nğŸ“Š PART 1: Create Sample Data | Ø¥Ù†Ø´Ø§Ø¡ Ø¨ÙŠØ§Ù†Ø§Øª Ù†Ù…ÙˆØ°Ø¬ÙŠØ©")print("-" * 70)np.random.seed(42)n = 100df = pd.DataFrame({    'category': ['A', 'B', 'C', 'D'] * 25,    'value': np.random.randn(n) * 10 + 50,    'time': pd.date_range('2024-01-01', periods=n, freq='D'),    'x': np.random.randn(n),    'y': np.random.randn(n) * 2 + np.random.randn(n)})
```

## Step 2

**What we're doing:** This section covers step 2.

**Why it matters:** Understanding this helps you master the concepts.


```python
print("Sample data created:")print(df.head())# ============================================================================# PART 2: MATPLOTLIB CHART TYPES | Ø£Ù†ÙˆØ§Ø¹ Ù…Ø®Ø·Ø·Ø§Øª Matplotlib# ============================================================================print("\n" + "=" * 70)print("PART 2: Matplotlib Chart Types | Ø£Ù†ÙˆØ§Ø¹ Ù…Ø®Ø·Ø·Ø§Øª Matplotlib")print("=" * 70)# Line Plotprint("\nâœ… Example 1: Line Plot (Trends Over Time)")print("-" * 70)plt.figure(figsize=(10, 5))plt.plot(df['time'][:30], df['value'][:30], marker='o', linewidth=2, markersize=4)plt.xlabel('Time')plt.ylabel('Value')plt.title('Line Plot: Value Over Time')plt.xticks(rotation=45)plt.grid(True, alpha=0.3)plt.tight_layout()plt.show()print("ğŸ’¡ Use for: Trends over time, continuous data")# Bar Chartprint("\nâœ… Example 2: Bar Chart (Categorical Comparison)")print("-" * 70)category_means = df.groupby('category')['value'].mean()plt.figure(figsize=(8, 5))category_means.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'orange'])plt.xlabel('Category')
```

## Step 3

**What we're doing:** This section covers step 3.

**Why it matters:** Understanding this helps you master the concepts.


```python
plt.ylabel('Average Value')plt.title('Bar Chart: Average Value by Category')plt.xticks(rotation=0)plt.grid(True, alpha=0.3, axis='y')plt.tight_layout()plt.show()print("ğŸ’¡ Use for: Comparing categories, discrete data")# Scatter Plotprint("\nâœ… Example 3: Scatter Plot (Relationships)")print("-" * 70)plt.figure(figsize=(8, 6))plt.scatter(df['x'], df['y'], alpha=0.6, s=50)plt.xlabel('X Variable')plt.ylabel('Y Variable')plt.title('Scatter Plot: Relationship Between X and Y')plt.grid(True, alpha=0.3)plt.tight_layout()plt.show()print("ğŸ’¡ Use for: Relationships, correlations, patterns")# Histogramprint("\nâœ… Example 4: Histogram (Distribution)")print("-" * 70)plt.figure(figsize=(8, 5))plt.hist(df['value'], bins=20, edgecolor='black', alpha=0.7)plt.xlabel('Value')plt.ylabel('Frequency')plt.title('Histogram: Value Distribution')plt.grid(True, alpha=0.3, axis='y')
```

## Step 4

```python
plt.tight_layout()plt.show()print("ğŸ’¡ Use for: Distributions, frequency, spread")# ============================================================================# PART 3: SEABORN CHART TYPES | Ø£Ù†ÙˆØ§Ø¹ Ù…Ø®Ø·Ø·Ø§Øª Seaborn# ============================================================================print("\n" + "=" * 70)print("PART 3: Seaborn Chart Types | Ø£Ù†ÙˆØ§Ø¹ Ù…Ø®Ø·Ø·Ø§Øª Seaborn")print("=" * 70)# Box Plotprint("\nâœ… Example 5: Box Plot (Statistical Summary)")print("-" * 70)plt.figure(figsize=(8, 5))sns.boxplot(data=df, x='category', y='value')plt.title('Box Plot: Value Distribution by Category')plt.tight_layout()plt.show()print("ğŸ’¡ Use for: Quartiles, outliers, distributions by category")# Heatmapprint("\nâœ… Example 6: Heatmap (Correlation Matrix)")print("-" * 70)correlation_data = df[['value', 'x', 'y']].corr()plt.figure(figsize=(8, 6))sns.heatmap(correlation_data, annot=True, cmap='coolwarm', center=0,             square=True, linewidths=1, cbar_kws={"shrink": 0.8})plt.title('Heatmap: Correlation Matrix')plt.tight_layout()
```
