```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("âœ… Libraries imported!")
print("\n" + "=" * 70)
print("Data Science Applications | ØªØ·Ø¨ÙŠÙ‚Ø§Øª Ø¹Ù„Ù… Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª")
print("=" * 70)

# ============================================================================
# PART 1: DATA SCIENCE LIFECYCLE IN PRACTICE | Ø¯ÙˆØ±Ø© Ø­ÙŠØ§Ø© Ø¹Ù„Ù… Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª ÙÙŠ Ø§Ù„Ù…Ù…Ø§Ø±Ø³Ø©
# ============================================================================
print("\nğŸ“Š PART 1: Data Science Lifecycle in Practice")
print("-" * 70)

print("""
The Data Science Lifecycle (from Example 1):

1. Problem Definition
   - What question are we trying to answer?
   - What data do we need?
   - What would success look like?

2. Data Collection
   - Gather relevant data
   - Check data quality
   - Understand data structure

3. Data Cleaning & Preparation
   - Handle missing values
   - Remove duplicates
   - Transform data

4. Analysis & Modeling
   - Explore data
   - Build models
   - Validate results

5. Results & Presentation
   - Visualize findings
   - Communicate insights
   - Make recommendations
""")

# ============================================================================
# PART 2: EXAMPLE PROJECT: SALES ANALYSIS | Ù…Ø«Ø§Ù„ Ø§Ù„Ù…Ø´Ø±ÙˆØ¹: ØªØ­Ù„ÙŠÙ„ Ø§Ù„Ù…Ø¨ÙŠØ¹Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
print("PART 2: Example Project - Sales Analysis | Ù…Ø«Ø§Ù„ Ø§Ù„Ù…Ø´Ø±ÙˆØ¹: ØªØ­Ù„ÙŠÙ„ Ø§Ù„Ù…Ø¨ÙŠØ¹Ø§Øª")
print("=" * 70)

# Step 1: Problem Definition
print("\nğŸ“‹ Step 1: Problem Definition")
print("-" * 70)
print("""
Problem: A retail store wants to understand sales patterns
Question: Which products sell best? What are the trends?
Goal: Provide insights to improve sales strategy
""")

# Step 2: Data Collection (simulated data)
print("\nğŸ“‹ Step 2: Data Collection")
print("-" * 70)

# Create sample sales data
sales_data = {
    'date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'product': ['A', 'B', 'A', 'C', 'A'],
    'sales': [100, 150, 120, 200, 110],
    'region': ['North', 'South', 'North', 'East', 'North']
}

df = pd.DataFrame(sales_data)
print("   Sample sales data collected:")
print(df.to_string(index=False))

# Step 3: Data Cleaning & Preparation
print("\nğŸ“‹ Step 3: Data Cleaning & Preparation")
print("-" * 70)

# Check for missing values
missing = df.isnull().sum()
print(f"   Missing values: {missing.sum()} (all clean!)")

# Check data types
print(f"   Data types:")
print(f"   {df.dtypes}")

# Step 4: Analysis
print("\nğŸ“‹ Step 4: Analysis")
print("-" * 70)

# Calculate statistics
total_sales = df['sales'].sum()
average_sales = df['sales'].mean()
best_product = df.groupby('product')['sales'].sum().idxmax()

print(f"   Total sales: ${total_sales}")
print(f"   Average sales: ${average_sales:.2f}")
print(f"   Best selling product: {best_product}")

# Product analysis
product_sales = df.groupby('product')['sales'].sum()
print(f"\n   Sales by product:")
for product, sales in product_sales.items():
    print(f"   - Product {product}: ${sales}")

# Step 5: Results & Presentation
print("\nğŸ“‹ Step 5: Results & Presentation")
print("-" * 70)
print("""
   Key Insights:
   1. Product A is the best seller (total: $330)
   2. Average sale is $136
   3. North region has most sales
   
   Recommendations:
   - Focus marketing on Product A
   - Expand Product A to other regions
   - Investigate why Product C has high individual sales
""")

# ============================================================================
# PART 3: INDUSTRY APPLICATIONS | Ø§Ù„ØªØ·Ø¨ÙŠÙ‚Ø§Øª Ø§Ù„ØµÙ†Ø§Ø¹ÙŠØ©
# ============================================================================
print("\n" + "=" * 70)
print("PART 3: Industry Applications | Ø§Ù„ØªØ·Ø¨ÙŠÙ‚Ø§Øª Ø§Ù„ØµÙ†Ø§Ø¹ÙŠØ©")
print("=" * 70)

applications = {
    "Healthcare": [
        "Patient diagnosis prediction",
        "Drug effectiveness analysis",
        "Hospital resource optimization"
    ],
    "Finance": [
        "Fraud detection",
        "Risk assessment",
        "Stock price prediction"
    ],
    "Retail": [
        "Customer segmentation",
        "Inventory optimization",
        "Sales forecasting"
    ],
    "Marketing": [
        "Campaign effectiveness",
        "Customer behavior analysis",
        "Price optimization"
    ]
}

for industry, uses in applications.items():
    print(f"\nğŸ¢ {industry}:")
    for use in uses:
        print(f"   - {use}")

# ============================================================================
# PART 4: PROJECT CHECKLIST | Ù‚Ø§Ø¦Ù…Ø© ÙØ­Øµ Ø§Ù„Ù…Ø´Ø±ÙˆØ¹
# ============================================================================
print("\n" + "=" * 70)
print("PART 4: Project Checklist | Ù‚Ø§Ø¦Ù…Ø© ÙØ­Øµ Ø§Ù„Ù…Ø´Ø±ÙˆØ¹")
print("=" * 70)

checklist = {
    "Planning": [
        "âœ… Clear problem definition",
        "âœ… Data requirements identified",
        "âœ… Success criteria defined"
    ],
    "Execution": [
        "âœ… Data collected and validated",
        "âœ… Data cleaned and prepared",
        "âœ… Analysis performed",
        "âœ… Results validated"
    ],
    "Presentation": [
        "âœ… Insights clearly communicated",
        "âœ… Visualizations created",
        "âœ… Recommendations provided"
    ]
}

for phase, items in checklist.items():
    print(f"\nğŸ“‹ {phase}:")
    for item in items:
        print(f"   {item}")

# ============================================================================
# SUMMARY | Ø§Ù„Ù…Ù„Ø®Øµ
# ============================================================================
print("\n" + "=" * 70)
print("Summary | Ø§Ù„Ù…Ù„Ø®Øµ")
print("=" * 70)
print("""
âœ… What you learned:
   1. Lifecycle in Practice: Complete workflow from problem to solution
   2. Example Project: Sales analysis demonstrating all steps
   3. Industry Applications: Real-world use cases
   4. Project Checklist: How to structure your projects

ğŸ¯ Key Takeaways:
   - Follow the lifecycle: Problem â†’ Data â†’ Clean â†’ Analyze â†’ Present
   - Start small: Simple projects build confidence
   - Document everything: Clear notebooks are essential
   - Practice: Each project makes you better

ğŸ“š Next Steps:
   - Unit 2: Data Cleaning (real-world data cleaning projects)
   - Unit 3: Data Visualization (present your findings)
   - Unit 4: Machine Learning (build predictive models)
   - Start your own project!
""")
print("âœ… Data science applications concepts understood!")
```
