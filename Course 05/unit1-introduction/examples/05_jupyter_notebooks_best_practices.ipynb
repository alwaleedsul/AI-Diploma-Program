**Common Student Questions:**
- **Q: Why use markdown cells?**
  - Answer: Markdown explains WHAT and WHY, code shows HOW
  - Example: Markdown explains "We're calculating average", code shows the calculation
  - Benefit: Makes notebooks readable and educational
  
- **Q: Why set random seeds?**
  - Answer: Ensures results are reproducible (same random numbers each run)
  - Example: `np.random.seed(42)` before generating random data
  - Benefit: Others get the same results when running your notebook
