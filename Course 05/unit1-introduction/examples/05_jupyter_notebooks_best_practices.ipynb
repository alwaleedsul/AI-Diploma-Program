## Step 2

**What we're doing:** This section covers step 2.

**Why it matters:** Understanding this helps you master the concepts.


````python
# ============================================================================
# PART 2: CODE ORGANIZATION | ØªÙ†Ø¸ÙŠÙ… Ø§Ù„ÙƒÙˆØ¯
# ============================================================================
print("\n" + "=" * 70)
print("PART 2: Code Organization | ØªÙ†Ø¸ÙŠÙ… Ø§Ù„ÙƒÙˆØ¯")
print("=" * 70)

# Example 1: Clear variable names
print("\nâœ… Example 1: Clear Variable Names")
print("-" * 70)

# âŒ Bad: Unclear names
x = [1, 2, 3, 4, 5]
y = sum(x) / len(x)

# âœ… Good: Clear, descriptive names
student_scores = [85, 90, 78, 92, 88]
average_score = sum(student_scores) / len(student_scores)

print(f"   Student scores: {student_scores}")
print(f"   Average score: {average_score}")
print("   âœ… Clear names make code self-documenting!")

# Example 2: Comments and explanations
print("\nâœ… Example 2: Comments and Explanations")
print("-" * 70)

# Calculate statistics with clear comments
data = [10, 20, 30, 40, 50]
````

## Step 3

**What we're doing:** This section covers step 3.

**Why it matters:** Understanding this helps you master the concepts.


````python
# Calculate mean (average)
mean_value = sum(data) / len(data)

# Calculate variance (spread of data)
variance = sum((x - mean_value) ** 2 for x in data) / len(data)

print(f"   Data: {data}")
print(f"   Mean: {mean_value}")
print(f"   Variance: {variance}")
print("   âœ… Comments explain WHY, not just WHAT")

# ============================================================================
# PART 3: REPRODUCIBILITY | Ø¥Ù…ÙƒØ§Ù†ÙŠØ© Ø§Ù„ØªÙƒØ±Ø§Ø±
# ============================================================================
print("\n" + "=" * 70)
print("PART 3: Reproducibility | Ø¥Ù…ÙƒØ§Ù†ÙŠØ© Ø§Ù„ØªÙƒØ±Ø§Ø±")
print("=" * 70)

import numpy as np
import random

# Example 1: Set random seeds
print("\nâœ… Example 1: Setting Random Seeds")
print("-" * 70)

# Set seed for reproducibility
np.random.seed(42)  # Same seed = same random numbers
random.seed(42)

# Generate random data
random_data_1 = np.random.rand(5)
random_data_2 = np.random.rand(5)
````

## Step 4

**What we're doing:** This section covers step 4.

**Why it matters:** Understanding this helps you master the concepts.


````python
print(f"   Random data (seed=42): {random_data_1}")
print("   âœ… Setting seed ensures same results every time!")

# Reset seed to show reproducibility
np.random.seed(42)
random_data_3 = np.random.rand(5)
print(f"   Same seed again: {random_data_3}")
print(f"   âœ… Results match: {np.array_equal(random_data_1, random_data_3)}")

# Example 2: Document dependencies
print("\nâœ… Example 2: Document Dependencies")
print("-" * 70)
print("""
   In your notebook, always include:
   
   ```python
   # Required libraries
   import pandas as pd
   import numpy as np
   import matplotlib.pyplot as plt
   
   # Check versions (optional but helpful)
   print(f"pandas version: {pd.__version__}")
   print(f"numpy version: {np.__version__}")
   ```
   
   âœ… This helps others install correct versions!
""")

# ============================================================================
# PART 4: NOTEBOOK BEST PRACTICES CHECKLIST | Ù‚Ø§Ø¦Ù…Ø© Ø£ÙØ¶Ù„ Ø§Ù„Ù…Ù…Ø§Ø±Ø³Ø§Øª
# ============================================================================
print("\n" + "=" * 70)
````
