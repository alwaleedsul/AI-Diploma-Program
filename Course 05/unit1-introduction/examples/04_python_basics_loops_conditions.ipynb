## Why Python Basics Matter | ŸÑŸÖÿßÿ∞ÿß ŸäŸáŸÖ ÿ£ÿ≥ÿßÿ≥Ÿäÿßÿ™ ÿ®ÿßŸäÿ´ŸàŸÜ

Python programming fundamentals are essential because:
- **Arithmetic Operations**: Calculate statistics, transform data, perform computations
- **Loops**: Process every row in a dataset, repeat operations, iterate over data
- **Conditions**: Filter data, make decisions, handle different cases
- **Together**: They form the foundation for all data science programming

**Common Student Questions:**
- **Q: Why do I need loops?**
  - Answer: Loops let you repeat code for every item in a list/dataset
  - Example: Calculate average of 1000 numbers (without loops, you'd write 1000 lines!)
  - In data science: Process every row, apply transformations to all data
  
- **Q: When do I use if/else?**
  - Answer: Use conditions to make decisions based on data values
  - Example: "If age > 18, then adult; else, child"
  - In data science: Filter data, handle missing values, categorize data

## Introduction

**Python programming fundamentals** are essential for data science, covering arithmetic operations, control flow with loops and conditions, and basic problem-solving. This notebook provides hands-on practice with working examples.

```python
# No external libraries needed for basic Python operations
print("‚úÖ Ready to learn Python basics!")
print("\n" + "=" * 70)
print("Python Programming: Basic Tasks | ÿßŸÑÿ®ÿ±ŸÖÿ¨ÿ© ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ŸÅŸä ÿ®ÿßŸäÿ´ŸàŸÜ")
print("=" * 70)

# ============================================================================
# PART 1: ARITHMETIC OPERATIONS | ÿßŸÑÿπŸÖŸÑŸäÿßÿ™ ÿßŸÑÿ≠ÿ≥ÿßÿ®Ÿäÿ©
# ============================================================================
print("\n" + "=" * 70)
print("PART 1: Arithmetic Operations | ÿßŸÑÿπŸÖŸÑŸäÿßÿ™ ÿßŸÑÿ≠ÿ≥ÿßÿ®Ÿäÿ©")
print("=" * 70)

# Basic arithmetic operations
a = 10
b = 3

print(f"\nGiven: a = {a}, b = {b}")
print(f"Addition (a + b): {a + b}")
print(f"Subtraction (a - b): {a - b}")
print(f"Multiplication (a * b): {a * b}")
print(f"Division (a / b): {a / b}")
print(f"Floor division (a // b): {a // b}  # Rounds down to nearest integer")
print(f"Modulo (a % b): {a % b}  # Returns remainder")
print(f"Exponentiation (a ** b): {a ** b}  # a to the power of b")

# Real-world example: Calculate average
numbers = [85, 90, 78, 92, 88]
total = sum(numbers)  # Sum all numbers
average = total / len(numbers)  # Divide by count
print(f"\nüìä Real-world example: Calculate average")
print(f"   Numbers: {numbers}")
print(f"   Sum: {total}")
print(f"   Count: {len(numbers)}")
print(f"   Average: {average}")

# ============================================================================
# PART 2: CONDITIONAL STATEMENTS | ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ¥ÿ±ÿ∑Ÿäÿ©
# ============================================================================
print("\n" + "=" * 70)
print("PART 2: Conditional Statements (if/elif/else) | ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ¥ÿ±ÿ∑Ÿäÿ©")
print("=" * 70)

# Simple if statement
age = 20
print(f"\nExample 1: Simple if statement")
print(f"   Age: {age}")
if age >= 18:
    print("   ‚úÖ Result: Adult (age >= 18)")

# if-else statement
score = 75
print(f"\nExample 2: if-else statement")
print(f"   Score: {score}")
if score >= 60:
    print("   ‚úÖ Result: Pass (score >= 60)")
else:
    print("   ‚ùå Result: Fail (score < 60)")

# if-elif-else statement
temperature = 25
print(f"\nExample 3: if-elif-else (multiple conditions)")
print(f"   Temperature: {temperature}¬∞C")
if temperature > 30:
    print("   üî• Result: Hot")
elif temperature > 20:
    print("   ‚òÄÔ∏è  Result: Warm")
elif temperature > 10:
    print("   üå§Ô∏è  Result: Mild")
else:
    print("   ‚ùÑÔ∏è  Result: Cold")

# Real-world example: Categorize data
data_values = [15, 25, 35, 5, 45]
print(f"\nüìä Real-world example: Categorize data values")
print(f"   Values: {data_values}")
for value in data_values:
    if value >= 30:
        category = "High"
    elif value >= 20:
        category = "Medium"
    else:
        category = "Low"
    print(f"   {value} ‚Üí {category}")

# ============================================================================
# PART 3: LOOPS | ÿßŸÑÿ≠ŸÑŸÇÿßÿ™
# ============================================================================
print("\n" + "=" * 70)
print("PART 3: Loops | ÿßŸÑÿ≠ŸÑŸÇÿßÿ™")
print("=" * 70)

# for loop: Iterate over a list
print("\nExample 1: for loop - iterate over list")
fruits = ["apple", "banana", "orange"]
print(f"   Fruits: {fruits}")
for fruit in fruits:
    print(f"   - {fruit}")

# for loop with range
print("\nExample 2: for loop - using range()")
print("   Counting from 1 to 5:")
for i in range(1, 6):  # range(1, 6) gives [1, 2, 3, 4, 5]
    print(f"   - Number {i}")

# while loop
print("\nExample 3: while loop - repeat until condition")
count = 0
print("   Counting from 0 to 4:")
while count < 5:
    print(f"   - Count: {count}")
    count += 1  # Increment count by 1

# Real-world example: Process data
print("\nüìä Real-world example: Process data with loop")
sales = [100, 150, 200, 180, 220]
total_sales = 0
print(f"   Daily sales: {sales}")
for sale in sales:
    total_sales += sale
    print(f"   - Day sales: ${sale}, Running total: ${total_sales}")
print(f"   ‚úÖ Total sales: ${total_sales}")

# List comprehension (advanced, but useful)
print("\nExample 4: List comprehension (advanced)")
numbers = [1, 2, 3, 4, 5]
squared = [x ** 2 for x in numbers]  # Square each number
print(f"   Original: {numbers}")
print(f"   Squared: {squared}")

# ============================================================================
# PART 4: COMBINING LOOPS AND CONDITIONS | ÿØŸÖÿ¨ ÿßŸÑÿ≠ŸÑŸÇÿßÿ™ ŸàÿßŸÑÿ¥ÿ±Ÿàÿ∑
# ============================================================================
print("\n" + "=" * 70)
print("PART 4: Combining Loops and Conditions | ÿØŸÖÿ¨ ÿßŸÑÿ≠ŸÑŸÇÿßÿ™ ŸàÿßŸÑÿ¥ÿ±Ÿàÿ∑")
print("=" * 70)

# Filter data using loop + condition
print("\nüìä Real-world example: Filter data")
scores = [45, 78, 92, 56, 88, 67, 95, 42]
print(f"   All scores: {scores}")
print("   Finding passing scores (>= 60):")
passing_scores = []
for score in scores:
    if score >= 60:
        passing_scores.append(score)
        print(f"   ‚úÖ {score} - Pass")
    else:
        print(f"   ‚ùå {score} - Fail")
print(f"   Passing scores: {passing_scores}")

# ============================================================================
# SUMMARY | ÿßŸÑŸÖŸÑÿÆÿµ
# ============================================================================
print("\n" + "=" * 70)
print("Summary | ÿßŸÑŸÖŸÑÿÆÿµ")
print("=" * 70)
print("""
‚úÖ What you learned:
   1. Arithmetic Operations: +, -, *, /, //, %, **
   2. Conditional Statements: if, elif, else (make decisions)
   3. Loops: for, while (repeat code)
   4. Combining loops and conditions (filter/process data)

üéØ Key Takeaways:
   - Arithmetic: Calculate values, transform data
   - Conditions: Make decisions, filter data
   - Loops: Process multiple items, repeat operations
   - Together: Foundation for all data science programming

üìö Next Steps:
   - Example 5: Jupyter Notebooks Best Practices
   - Example 2: Pandas & NumPy Basics (uses loops and conditions!)
   - Unit 2: Data Cleaning (heavily uses loops and conditions)
""")
print("‚úÖ Python basics concepts understood!")
```
