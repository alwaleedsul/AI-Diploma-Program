## Case Study 1: Autonomous Vacuum Cleaner Agent
## Ø¯Ø±Ø§Ø³Ø© Ø­Ø§Ù„Ø© 1: ÙˆÙƒÙŠÙ„ Ù…ÙƒÙ†Ø³Ø© ÙƒÙ‡Ø±Ø¨Ø§Ø¦ÙŠØ© Ø°Ø§ØªÙŠØ©

Let's analyze a simple autonomous vacuum cleaner as an intelligent agent using the PEAS framework.


```python
# Case Study 1: Autonomous Vacuum Cleaner Agent
# PEAS Framework Analysis

print("=== PEAS Framework Analysis ===")
print("\nPerformance Measure:")
print("- Cleanliness of floors")
print("- Time to complete cleaning")
print("- Energy efficiency")
print("- Battery life")

print("\nEnvironment:")
print("- Type: Discrete, Partially Observable, Sequential")
print("- Characteristics: Grid-based floor layout, dynamic (people, pets)")
print("- Agents: Single agent (vacuum cleaner)")

print("\nActuators:")
print("- Movement motors (forward, backward, turn)")
print("- Vacuum motor (on/off)")
print("- Brush mechanism")

print("\nSensors:")
print("- Dirt sensors")
print("- Bump sensors (collision detection)")
print("- Battery level sensor")
print("- Camera/LIDAR for mapping (in advanced models)")
```

```python
# Simple Reflex Vacuum Agent Implementation
class SimpleVacuumAgent:
    """Simple reflex-based vacuum cleaner agent"""
    def __init__(self):
        self.location = (0, 0)
        self.battery = 100
        self.cleaned = set()
    
    def sense(self, environment):
        """Sense the environment"""
        x, y = self.location
        return {
            'dirt': environment.get((x, y), False),
            'battery': self.battery,
            'location': (x, y)
        }
    
    def act(self, perception):
        """Simple reflex: if dirty, clean; else move"""
        if perception['battery'] <= 10:
            return 'return_to_charger'
        elif perception['dirt']:
            self.cleaned.add(self.location)
            self.battery -= 1
            return 'clean'
        else:
            self.battery -= 1
            return 'move_forward'

# Example usage
agent = SimpleVacuumAgent()
environment = {(0, 0): True, (1, 0): False, (2, 0): True}

print("=== Simple Vacuum Agent Decision-Making ===")
for step in range(5):
    perception = agent.sense(environment)
    action = agent.act(perception)
    print(f"Step {step + 1}: Location {perception['location']}, Dirt: {perception['dirt']}, Action: {action}")
```
