```python
"""
Unit 2 - Example 1: Breadth-First Search (BFS)
Ø§Ù„ÙˆØ­Ø¯Ø© 2 - Ù…Ø«Ø§Ù„ 1: Ø§Ù„Ø¨Ø­Ø« Ø¨Ø§Ù„Ø¹Ø±Ø¶ Ø£ÙˆÙ„Ø§Ù‹

This example demonstrates:
1. BFS algorithm implementation
2. Finding shortest path in a graph
3. Visualizing the search process
"""

from collections import deque

print("=" * 60)
print("Example 1: Breadth-First Search (BFS)")
print("Ù…Ø«Ø§Ù„ 1: Ø§Ù„Ø¨Ø­Ø« Ø¨Ø§Ù„Ø¹Ø±Ø¶ Ø£ÙˆÙ„Ø§Ù‹")
print("=" * 60)

# Graph representation
# ØªÙ…Ø«ÙŠÙ„ Ø§Ù„Ø±Ø³Ù… Ø§Ù„Ø¨ÙŠØ§Ù†ÙŠ
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

def bfs(graph, start, goal):
    """
    Breadth-First Search algorithm.
    Ø®ÙˆØ§Ø±Ø²Ù…ÙŠØ© Ø§Ù„Ø¨Ø­Ø« Ø¨Ø§Ù„Ø¹Ø±Ø¶ Ø£ÙˆÙ„Ø§Ù‹.
    
    Args:
        graph: Dictionary representing the graph
        start: Starting node
        goal: Target node
    
    Returns:
        Path from start to goal, or None if not found
    """
    # Queue for BFS
    queue = deque([(start, [start])])
    visited = set([start])
    
    print(f"\nSearching for path from {start} to {goal}...")
    print(f"Ø§Ù„Ø¨Ø­Ø« Ø¹Ù† Ù…Ø³Ø§Ø± Ù…Ù† {start} Ø¥Ù„Ù‰ {goal}...")
    
    while queue:
        current, path = queue.popleft()
        
        print(f"  Visiting: {current}, Path: {' -> '.join(path)}")
        
        if current == goal:
            print(f"\nâœ“ Goal found! Path: {' -> '.join(path)}")
            return path
        
        # Explore neighbors
        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    
    print(f"\nâœ— No path found from {start} to {goal}")
    return None

# Test BFS
print("\n" + "=" * 60)
print("Testing BFS Algorithm")
print("Ø§Ø®ØªØ¨Ø§Ø± Ø®ÙˆØ§Ø±Ø²Ù…ÙŠØ© BFS")
print("=" * 60)

# Test case 1
path1 = bfs(graph, 'A', 'F')
print(f"\nResult: {path1}")

# Test case 2
print("\n" + "-" * 60)
path2 = bfs(graph, 'D', 'C')
print(f"\nResult: {path2}")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ØªÙ… Ø¥ÙƒÙ…Ø§Ù„ Ø§Ù„Ù…Ø«Ø§Ù„ Ø¨Ù†Ø¬Ø§Ø­!")
print("=" * 60)
```
