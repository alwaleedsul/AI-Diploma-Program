```python
"""
Unit 4 - Example 1: Simple Perceptron
Ø§Ù„ÙˆØ­Ø¯Ø© 4 - Ù…Ø«Ø§Ù„ 1: Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ† Ø§Ù„Ø¨Ø³ÙŠØ·

This example demonstrates:
1. Perceptron implementation
2. Training a perceptron
3. Binary classification
4. Decision boundary visualization
"""

import numpy as np
import matplotlib.pyplot as plt

print("=" * 60)
print("Example 1: Simple Perceptron")
print("Ù…Ø«Ø§Ù„ 1: Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ† Ø§Ù„Ø¨Ø³ÙŠØ·")
print("=" * 60)

class SimplePerceptron:
    """
    Simple Perceptron for binary classification.
    Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ† Ø¨Ø³ÙŠØ· Ù„Ù„ØªØµÙ†ÙŠÙ Ø§Ù„Ø«Ù†Ø§Ø¦ÙŠ.
    """
    
    def __init__(self, learning_rate=0.1, epochs=100):
        """
        Initialize perceptron.
        ØªÙ‡ÙŠØ¦Ø© Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†.
        
        Args:
            learning_rate: Learning rate (default: 0.1)
            epochs: Number of training iterations (default: 100)
        """
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
    
    def activation(self, x):
        """
        Step activation function.
        Ø¯Ø§Ù„Ø© Ø§Ù„ØªÙ†Ø´ÙŠØ· Ø§Ù„Ø®Ø·ÙˆØ©.
        
        Returns 1 if x >= 0, else 0
        """
        return 1 if x >= 0 else 0
    
    def fit(self, X, y):
        """
        Train the perceptron.
        ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†.
        
        Args:
            X: Input features (n_samples, n_features)
            y: Target labels (0 or 1)
        """
        n_samples, n_features = X.shape
        
        # Initialize weights and bias
        # ØªÙ‡ÙŠØ¦Ø© Ø§Ù„Ø£ÙˆØ²Ø§Ù† ÙˆØ§Ù„Ø§Ù†Ø­ÙŠØ§Ø²
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        print("\nTraining Perceptron...")
        print("ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†...")
        
        for epoch in range(self.epochs):
            errors = 0
            for i in range(n_samples):
                # Forward pass
                # Ø§Ù„Ù…Ø±ÙˆØ± Ø§Ù„Ø£Ù…Ø§Ù…ÙŠ
                linear_output = np.dot(X[i], self.weights) + self.bias
                prediction = self.activation(linear_output)
                
                # Update weights if prediction is wrong
                # ØªØ­Ø¯ÙŠØ« Ø§Ù„Ø£ÙˆØ²Ø§Ù† Ø¥Ø°Ø§ ÙƒØ§Ù†Øª Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª Ø®Ø§Ø·Ø¦Ø©
                error = y[i] - prediction
                if error != 0:
                    errors += 1
                    self.weights += self.learning_rate * error * X[i]
                    self.bias += self.learning_rate * error
            
            if errors == 0:
                print(f"  Converged at epoch {epoch + 1}")
                print(f"  ØªÙ‚Ø§Ø±Ø¨ ÙÙŠ Ø§Ù„Ø¹ØµØ± {epoch + 1}")
                break
            
            if (epoch + 1) % 10 == 0:
                print(f"  Epoch {epoch + 1}: {errors} errors")
        
        print(f"\nTraining complete!")
        print(f"ØªÙ… Ø§Ù„ØªØ¯Ø±ÙŠØ¨!")
        print(f"Final weights: {self.weights}")
        print(f"Final bias: {self.bias}")
    
    def predict(self, X):
        """
        Make predictions.
        Ø¹Ù…Ù„ Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª.
        """
        predictions = []
        for i in range(len(X)):
            linear_output = np.dot(X[i], self.weights) + self.bias
            predictions.append(self.activation(linear_output))
        return np.array(predictions)

# Example: AND Gate
# Ù…Ø«Ø§Ù„: Ø¨ÙˆØ§Ø¨Ø© AND
print("\n" + "=" * 60)
print("Example: AND Gate Classification")
print("Ù…Ø«Ø§Ù„: ØªØµÙ†ÙŠÙ Ø¨ÙˆØ§Ø¨Ø© AND")
print("=" * 60)

# Training data for AND gate
# Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨ Ù„Ø¨ÙˆØ§Ø¨Ø© AND
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND gate output

print("\nTraining Data:")
print("Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨:")
print("Input | Output")
print("------|-------")
for i in range(len(X)):
    print(f"{X[i]}  |   {y[i]}")

# Create and train perceptron
# Ø¥Ù†Ø´Ø§Ø¡ ÙˆØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†
perceptron = SimplePerceptron(learning_rate=0.1, epochs=100)
perceptron.fit(X, y)

# Test the perceptron
# Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†
print("\n" + "=" * 60)
print("Testing Perceptron")
print("Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ø¨ÙŠØ±Ø³Ø¨ØªØ±ÙˆÙ†")
print("=" * 60)

predictions = perceptron.predict(X)
print("\nPredictions:")
print("Ø§Ù„ØªÙˆÙ‚Ø¹Ø§Øª:")
print("Input | Expected | Predicted | Correct")
print("------|----------|-----------|--------")
for i in range(len(X)):
    correct = "âœ“" if predictions[i] == y[i] else "âœ—"
    print(f"{X[i]}  |    {y[i]}     |     {predictions[i]}     |   {correct}")

accuracy = np.mean(predictions == y) * 100
print(f"\nAccuracy: {accuracy}%")
print(f"Ø§Ù„Ø¯Ù‚Ø©: {accuracy}%")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ØªÙ… Ø¥ÙƒÙ…Ø§Ù„ Ø§Ù„Ù…Ø«Ø§Ù„ Ø¨Ù†Ø¬Ø§Ø­!")
print("=" * 60)
```
