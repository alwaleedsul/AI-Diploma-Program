```python
"""
Unit 3 - Example 1: RNN Basics
ÿßŸÑŸàÿ≠ÿØÿ© 3 - ŸÖÿ´ÿßŸÑ 1: ÿ£ÿ≥ÿßÿ≥Ÿäÿßÿ™ ÿßŸÑÿ¥ÿ®ŸÉÿßÿ™ ÿßŸÑÿπÿµÿ®Ÿäÿ© ÿßŸÑÿ™ŸÉÿ±ÿßÿ±Ÿäÿ©

This example demonstrates:
1. Understanding RNN architecture
2. Simple RNN implementation
3. Sequence processing
4. Basic RNN operations
"""

print("=" * 60)
print("Example 1: RNN Basics")
print("ŸÖÿ´ÿßŸÑ 1: ÿ£ÿ≥ÿßÿ≥Ÿäÿßÿ™ ÿßŸÑÿ¥ÿ®ŸÉÿßÿ™ ÿßŸÑÿπÿµÿ®Ÿäÿ© ÿßŸÑÿ™ŸÉÿ±ÿßÿ±Ÿäÿ©")
print("=" * 60)

# Simple RNN demonstration
# ÿπÿ±ÿ∂ ÿ®ÿ≥Ÿäÿ∑ ŸÑŸÑÿ¥ÿ®ŸÉÿ© ÿßŸÑÿπÿµÿ®Ÿäÿ© ÿßŸÑÿ™ŸÉÿ±ÿßÿ±Ÿäÿ©
print("\nRNN Architecture Overview:")
print("ŸÜÿ∏ÿ±ÿ© ÿπÿßŸÖÿ© ÿπŸÑŸâ ŸáŸäŸÉŸÑ RNN:")
print("-" * 60)

rnn_structure = """
Recurrent Neural Network (RNN):
‚îú‚îÄ‚îÄ Input: Sequence of data (x‚ÇÅ, x‚ÇÇ, ..., x‚Çú)
‚îú‚îÄ‚îÄ Hidden State: Maintains memory (h‚ÇÄ, h‚ÇÅ, ..., h‚Çú)
‚îú‚îÄ‚îÄ Output: Sequence of predictions (y‚ÇÅ, y‚ÇÇ, ..., y‚Çú)
‚îî‚îÄ‚îÄ Key Feature: Hidden state passes information forward
"""

print(rnn_structure)

# Mathematical representation
print("\nMathematical Representation:")
print("ÿßŸÑÿ™ŸÖÿ´ŸäŸÑ ÿßŸÑÿ±Ÿäÿßÿ∂Ÿä:")
print("-" * 60)

math_formula = """
h‚Çú = tanh(W‚Çï‚Çï ¬∑ h‚Çú‚Çã‚ÇÅ + W‚Çì‚Çï ¬∑ x‚Çú + b)
y‚Çú = W‚Çï·µß ¬∑ h‚Çú + b·µß

Where:
- h‚Çú: hidden state at time t
- x‚Çú: input at time t
- y‚Çú: output at time t
- W: weight matrices
- b: bias terms
"""

print(math_formula)

# Simple RNN implementation concept
print("\nSimple RNN Implementation Concept:")
print("ŸÖŸÅŸáŸàŸÖ ÿ™ŸÜŸÅŸäÿ∞ RNN ÿßŸÑÿ®ÿ≥Ÿäÿ∑:")
print("-" * 60)

code_concept = """
# Pseudo-code for RNN
# ŸÉŸàÿØ ÿ¥ÿ®Ÿá ÿ®ÿ±ŸÖÿ¨Ÿä ŸÑŸÄ RNN

def rnn_step(hidden_state, input_data, weights):
    # Combine previous hidden state with current input
    combined = weights['hh'] @ hidden_state + weights['xh'] @ input_data
    # Apply activation function
    new_hidden = tanh(combined + weights['bias'])
    # Generate output
    output = weights['hy'] @ new_hidden
    return new_hidden, output
"""

print(code_concept)

# Visualization
try:
    import matplotlib.pyplot as plt
    import numpy as np
    
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Draw RNN unrolled structure
    time_steps = 4
    x_positions = [1, 3, 5, 7]
    
    for i, x in enumerate(x_positions):
        # Input
        circle_input = plt.Circle((x, 6), 0.3, color='lightblue', ec='black', linewidth=2)
        ax.add_patch(circle_input)
        ax.text(x, 6, f'x{i+1}', ha='center', va='center', fontsize=10, fontweight='bold')
        
        # Hidden state
        rect_hidden = plt.Rectangle((x-0.4, 3.5), 0.8, 1, 
                                   color='lightgreen', ec='black', linewidth=2)
        ax.add_patch(rect_hidden)
        ax.text(x, 4, f'h{i+1}', ha='center', va='center', fontsize=10, fontweight='bold')
        
        # Output
        circle_output = plt.Circle((x, 1.5), 0.3, color='lightcoral', ec='black', linewidth=2)
        ax.add_patch(circle_output)
        ax.text(x, 1.5, f'y{i+1}', ha='center', va='center', fontsize=10, fontweight='bold')
        
        # Arrows forward
        if i < len(x_positions) - 1:
            # Input to hidden
            ax.arrow(x+0.3, 5.7, 1.4, -1.2, head_width=0.15, head_length=0.1, 
                    fc='black', ec='black', linewidth=1.5)
            # Hidden to next hidden
            ax.arrow(x+0.4, 4, 1.2, 0, head_width=0.2, head_length=0.15, 
                    fc='green', ec='green', linewidth=2, alpha=0.7)
            # Hidden to output
            ax.arrow(x, 3.5, 0, -1.5, head_width=0.15, head_length=0.1, 
                    fc='black', ec='black', linewidth=1.5)
    
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 7)
    ax.set_title('RNN Unrolled Structure | ŸáŸäŸÉŸÑ RNN ÿ∫Ÿäÿ± ÿßŸÑŸÖÿ∑ŸàŸä', fontsize=16, pad=20, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    print("‚úÖ RNN architecture diagram displayed")
except ImportError:
    print("Note: Install matplotlib for architecture diagrams")
    print("ŸÖŸÑÿßÿ≠ÿ∏ÿ©: ŸÇŸÖ ÿ®ÿ™ÿ´ÿ®Ÿäÿ™ matplotlib ŸÑŸÖÿÆÿ∑ÿ∑ÿßÿ™ ÿßŸÑŸáŸäŸÉŸÑ")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ÿ™ŸÖ ÿ•ŸÉŸÖÿßŸÑ ÿßŸÑŸÖÿ´ÿßŸÑ ÿ®ŸÜÿ¨ÿßÿ≠!")
print("=" * 60)
print("\nNote: For actual implementation, use TensorFlow/Keras:")
print("from tensorflow.keras.layers import SimpleRNN, LSTM, GRU")
```
