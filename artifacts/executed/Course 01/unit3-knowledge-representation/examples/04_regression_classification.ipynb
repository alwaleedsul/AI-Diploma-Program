```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
import matplotlib.pyplot as plt

print("=== Regression vs Classification ===")
print("\nRegression:")
print("  - Predicts continuous values (numbers)")
print("  - Examples: Price, Temperature, Age")
print("  - Output: Real numbers")
print("  - Metrics: MSE, RMSE, MAE, RÂ²")

print("\nClassification:")
print("  - Predicts discrete categories (classes)")
print("  - Examples: Spam/Not Spam, Disease/Healthy, Cat/Dog")
print("  - Output: Class labels")
print("  - Metrics: Accuracy, Precision, Recall, F1-score")
```

## Example 1: Regression - House Price Prediction


```python
# Regression Example: Predicting house prices (continuous value)
np.random.seed(42)
house_sizes = np.random.randint(1000, 3000, 100)
house_prices = house_sizes * 150 + np.random.normal(0, 20000, 100)

X_reg = house_sizes.reshape(-1, 1)
y_reg = house_prices

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Train regression model
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)

# Predictions
y_pred = reg_model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("=== Regression Model Results ===")
print(f"Model: Linear Regression")
print(f"Mean Squared Error (MSE): {mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"RÂ² Score: {reg_model.score(X_test, y_test):.4f}")
print(f"\nExample prediction:")
print(f"  House size: 2000 sqft")
print(f"  Predicted price: ${reg_model.predict([[2000]])[0]:,.2f}")
```

## Example 2: Classification - Email Spam Detection


```python
# Classification Example: Spam detection (discrete classes)
np.random.seed(42)
# Features: word count, exclamation marks, length
word_count = np.random.randint(10, 200, 100)
exclamation = np.random.randint(0, 10, 100)
length = np.random.randint(50, 1000, 100)

# Create labels: 1 = spam, 0 = ham
# Spam tends to have more exclamation marks and shorter length
labels = ((exclamation > 5) | (length < 200)).astype(int)

X_clf = np.column_stack([word_count, exclamation, length])
y_clf = labels

# Split data
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

# Train classification model
clf_model = LogisticRegression()
clf_model.fit(X_train_clf, y_train_clf)

# Predictions
y_pred_clf = clf_model.predict(X_test_clf)

# Evaluate
accuracy = accuracy_score(y_test_clf, y_pred_clf)

print("=== Classification Model Results ===")
print(f"Model: Logistic Regression")
print(f"Accuracy: {accuracy:.4f}")
print(f"\nClassification Report:")
print(classification_report(y_test_clf, y_pred_clf, target_names=['Ham', 'Spam']))
```
