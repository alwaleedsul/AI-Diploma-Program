```python


# Visualization: Progress Chart
# ÿ™ÿµŸàÿ±: ŸÖÿÆÿ∑ÿ∑ ÿßŸÑÿ™ŸÇÿØŸÖ
try:
    import matplotlib.pyplot as plt
    
    # Example progress visualization
    steps = ['Step 1', 'Step 2', 'Step 3', 'Step 4']
    progress = [25, 50, 75, 100]
    
    plt.figure(figsize=(10, 6))
    plt.barh(steps, progress, color='green', alpha=0.7)
    plt.xlabel('Progress % | ÿßŸÑŸÜÿ≥ÿ®ÿ© ÿßŸÑŸÖÿ¶ŸàŸäÿ© ŸÑŸÑÿ™ŸÇÿØŸÖ', fontsize=12)
    plt.title('Learning Progress | ÿ™ŸÇÿØŸÖ ÿßŸÑÿ™ÿπŸÑŸÖ', fontsize=14, pad=20)
    plt.xlim(0, 100)
    for i, v in enumerate(progress):
        plt.text(v + 2, i, f'{v}%', va='center', fontsize=11)
    plt.tight_layout()
    plt.show()
    print("‚úÖ Progress visualization displayed")
except ImportError:
    print("Note: Install matplotlib for visualizations")
    print("ŸÖŸÑÿßÿ≠ÿ∏ÿ©: ŸÇŸÖ ÿ®ÿ™ÿ´ÿ®Ÿäÿ™ matplotlib ŸÑŸÑÿ™ÿµŸàÿ±ÿßÿ™")

"""
Unit 1 - Example 1: Introduction to Artificial Intelligence
ÿßŸÑŸàÿ≠ÿØÿ© 1 - ŸÖÿ´ÿßŸÑ 1: ŸÖŸÇÿØŸÖÿ© ŸÅŸä ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä

This example demonstrates:
1. Basic AI concepts and definitions
2. Types of AI systems
3. Real-world AI applications
4. Simple AI decision-making example
"""

print("=" * 60)
print("Example 1: Introduction to Artificial Intelligence")
print("ŸÖÿ´ÿßŸÑ 1: ŸÖŸÇÿØŸÖÿ© ŸÅŸä ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä")
print("=" * 60)

# 1. What is AI?
# ŸÖÿß ŸáŸà ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸäÿü
print("\n1. What is Artificial Intelligence?")
print("ŸÖÿß ŸáŸà ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸäÿü")
print("-" * 60)

ai_definition = """
Artificial Intelligence (AI) is the simulation of human intelligence 
in machines that are programmed to think and learn like humans.

ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä ŸáŸà ŸÖÿ≠ÿßŸÉÿßÿ© ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿ®ÿ¥ÿ±Ÿä ŸÅŸä ÿßŸÑÿ¢ŸÑÿßÿ™ ÿßŸÑŸÖÿ®ÿ±ŸÖÿ¨ÿ© 
ŸÑŸÑÿ™ŸÅŸÉŸäÿ± ŸàÿßŸÑÿ™ÿπŸÑŸÖ ŸÖÿ´ŸÑ ÿßŸÑÿ®ÿ¥ÿ±.
"""

print(ai_definition)

# 2. Types of AI Systems
# ÿ£ŸÜŸàÿßÿπ ÿ£ŸÜÿ∏ŸÖÿ© ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä
print("\n2. Types of AI Systems")
print("ÿ£ŸÜŸàÿßÿπ ÿ£ŸÜÿ∏ŸÖÿ© ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä")
print("-" * 60)

ai_types = {
    "Narrow AI (Weak AI)": {
        "description": "AI designed for specific tasks", "examples": ["Siri", "Google Translate", "Image Recognition"],
        "arabic": "ÿ∞ŸÉÿßÿ° ÿßÿµÿ∑ŸÜÿßÿπŸä ÿ∂ŸäŸÇ (ÿ∂ÿπŸäŸÅ) - ŸÖÿµŸÖŸÖ ŸÑŸÖŸáÿßŸÖ ŸÖÿ≠ÿØÿØÿ©"
    },
    "General AI (Strong AI)": {
        "description": "AI with human-level intelligence across all domains",
        "examples": ["Not yet achieved", "Theoretical"],
        "arabic": "ÿ∞ŸÉÿßÿ° ÿßÿµÿ∑ŸÜÿßÿπŸä ÿπÿßŸÖ (ŸÇŸàŸä) - ÿ∞ŸÉÿßÿ° ÿπŸÑŸâ ŸÖÿ≥ÿ™ŸàŸâ ÿßŸÑÿ®ÿ¥ÿ±"
    },
    "Superintelligent AI": {
        "description": "AI that surpasses human intelligence",
        "examples": ["Future concept", "Theoretical"],
        "arabic": "ÿ∞ŸÉÿßÿ° ÿßÿµÿ∑ŸÜÿßÿπŸä ŸÅÿßÿ¶ŸÇ - Ÿäÿ™ÿ¨ÿßŸàÿ≤ ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿ®ÿ¥ÿ±Ÿä"
    }
}

for ai_type, info in ai_types.items():
    print(f"\n{ai_type}:")
    print(f"  {info['arabic']}")
    print(f"  Description: {info['description']}")
    print(f"  Examples: {', '.join(info['examples'])}")

# 3. Simple AI Decision-Making Example
# ŸÖÿ´ÿßŸÑ ÿ®ÿ≥Ÿäÿ∑ ÿπŸÑŸâ ÿßÿ™ÿÆÿßÿ∞ ÿßŸÑŸÇÿ±ÿßÿ± ÿ®ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä
print("\n" + "=" * 60)
print("3. Simple AI Decision-Making Example")
print("ŸÖÿ´ÿßŸÑ ÿ®ÿ≥Ÿäÿ∑ ÿπŸÑŸâ ÿßÿ™ÿÆÿßÿ∞ ÿßŸÑŸÇÿ±ÿßÿ± ÿ®ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä")
print("=" * 60)

def simple_ai_decision(temperature, humidity, time_of_day):
    """
    Simple rule-based AI system for weather-based activity recommendation.
    ŸÜÿ∏ÿßŸÖ ÿ∞ŸÉÿßÿ° ÿßÿµÿ∑ŸÜÿßÿπŸä ÿ®ÿ≥Ÿäÿ∑ ŸÇÿßÿ¶ŸÖ ÿπŸÑŸâ ÿßŸÑŸÇŸàÿßÿπÿØ ŸÑÿ™ŸàÿµŸäÿ© ŸÜÿ¥ÿßÿ∑ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿßŸÑÿ∑ŸÇÿ≥.
    
    Args:
        temperature: Temperature in Celsius
        humidity: Humidity percentage (0-100)
        time_of_day: 'morning', 'afternoon', or 'evening'
    
    Returns:
        Recommended activity
    """
    # Decision rules / ŸÇŸàÿßÿπÿØ ÿßŸÑŸÇÿ±ÿßÿ±
    if temperature > 25 and humidity < 60:
        if time_of_day == 'morning':
            return "Go for a jog in the park | ÿßÿ∞Ÿáÿ® ŸÑŸÑÿ¨ÿ±Ÿä ŸÅŸä ÿßŸÑÿ≠ÿØŸäŸÇÿ©"
        elif time_of_day == 'afternoon':
            return "Go swimming | ÿßÿ∞Ÿáÿ® ŸÑŸÑÿ≥ÿ®ÿßÿ≠ÿ©"
        else:
            return "Take an evening walk | ÿßÿ∞Ÿáÿ® ŸÑŸÜÿ≤Ÿáÿ© ŸÖÿ≥ÿßÿ¶Ÿäÿ©"
    
    elif temperature < 15:
        return "Stay indoors, read a book | ÿßÿ®ŸÇ ŸÅŸä ÿßŸÑÿØÿßÿÆŸÑ ŸàÿßŸÇÿ±ÿ£ ŸÉÿ™ÿßÿ®ÿßŸã"
    
    elif humidity > 80:
        return "Stay indoors, watch a movie | ÿßÿ®ŸÇ ŸÅŸä ÿßŸÑÿØÿßÿÆŸÑ Ÿàÿ¥ÿßŸáÿØ ŸÅŸäŸÑŸÖÿßŸã"
    
    else:
        return "Moderate weather, any outdoor activity is fine | ÿ∑ŸÇÿ≥ ŸÖÿπÿ™ÿØŸÑÿå ÿ£Ÿä ŸÜÿ¥ÿßÿ∑ ÿÆÿßÿ±ÿ¨Ÿä ŸÖŸÜÿßÿ≥ÿ®"

# Test the AI system
# ÿßÿÆÿ™ÿ®ÿßÿ± ŸÜÿ∏ÿßŸÖ ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä
print("\nTesting AI Decision System:")
print("ÿßÿÆÿ™ÿ®ÿßÿ± ŸÜÿ∏ÿßŸÖ ÿßÿ™ÿÆÿßÿ∞ ÿßŸÑŸÇÿ±ÿßÿ±:")

test_cases = [
    (28, 50, 'morning'),
    (30, 45, 'afternoon'),
    (12, 70, 'evening'),
    (20, 85, 'afternoon')
]

for temp, hum, time in test_cases:
    decision = simple_ai_decision(temp, hum, time)
    print(f"\nTemperature: {temp}¬∞C, Humidity: {hum}%, Time: {time}")
    print(f"Recommendation: {decision}")

# 4. AI Applications in Real World
# ÿ™ÿ∑ÿ®ŸäŸÇÿßÿ™ ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä ŸÅŸä ÿßŸÑÿπÿßŸÑŸÖ ÿßŸÑÿ≠ŸÇŸäŸÇŸä
print("\n" + "=" * 60)
print("4. AI Applications in Real World")
print("ÿ™ÿ∑ÿ®ŸäŸÇÿßÿ™ ÿßŸÑÿ∞ŸÉÿßÿ° ÿßŸÑÿßÿµÿ∑ŸÜÿßÿπŸä ŸÅŸä ÿßŸÑÿπÿßŸÑŸÖ ÿßŸÑÿ≠ŸÇŸäŸÇŸä")
print("=" * 60)

applications = {
    "Healthcare": [
        "Medical image analysis", "Drug discovery",
        "Personalized treatment plans"
    ],
    "Transportation": [
        "Autonomous vehicles",
        "Traffic optimization",
        "Route planning"
    ],
    "Finance": [
        "Fraud detection",
        "Algorithmic trading",
        "Credit scoring"
    ],
    "Education": [
        "Personalized learning",
        "Automated grading",
        "Intelligent tutoring systems"
    ]
}

for domain, apps in applications.items():
    print(f"\n{domain}:")
    for app in apps:
        print(f"  - {app}")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ÿ™ŸÖ ÿ•ŸÉŸÖÿßŸÑ ÿßŸÑŸÖÿ´ÿßŸÑ ÿ®ŸÜÿ¨ÿßÿ≠!")
print("=" * 60)
```
