```python


# Visualization: Q-Table Heatmap
# ÿ™ÿµŸàÿ±: ÿÆÿ±Ÿäÿ∑ÿ© ÿ≠ÿ±ÿßÿ±Ÿäÿ© ŸÑÿ¨ÿØŸàŸÑ Q
try:
    import matplotlib.pyplot as plt
    import numpy as np
    
    # Example Q-table (from training above)
    q_table_example = np.array([[0.5, 0.8], [0.3, 0.9], [0.2, 0.7], [0.1, 0.6], [0.0, 0.0]])
    
    plt.figure(figsize=(8, 6))
    plt.imshow(q_table_example, cmap='YlOrRd', aspect='auto')
    plt.colorbar(label='Q-Value | ŸÇŸäŸÖÿ© Q')
    plt.xlabel('Actions | ÿßŸÑÿ•ÿ¨ÿ±ÿßÿ°ÿßÿ™ (Left, Right)', fontsize=12)
    plt.ylabel('States | ÿßŸÑÿ≠ÿßŸÑÿßÿ™', fontsize=12)
    plt.title('Q-Table Visualization | ÿ™ÿµŸàÿ± ÿ¨ÿØŸàŸÑ Q', fontsize=14, pad=20)
    plt.xticks([0, 1], ['Left | Ÿäÿ≥ÿßÿ±', 'Right | ŸäŸÖŸäŸÜ'])
    plt.yticks(range(5), [f'State {i}' for i in range(5)])
    plt.tight_layout()
    plt.show()
    print("‚úÖ Q-table heatmap displayed")
except ImportError:
    print("Note: Install matplotlib and numpy for Q-table visualization")
    print("ŸÖŸÑÿßÿ≠ÿ∏ÿ©: ŸÇŸÖ ÿ®ÿ™ÿ´ÿ®Ÿäÿ™ matplotlib Ÿà numpy ŸÑÿ™ÿµŸàÿ± ÿ¨ÿØŸàŸÑ Q")

"""
Unit 1 - Example 1: Markov Decision Process (MDP)
ÿßŸÑŸàÿ≠ÿØÿ© 1 - ŸÖÿ´ÿßŸÑ 1: ÿπŸÖŸÑŸäÿ© ÿßŸÑŸÇÿ±ÿßÿ± ÿßŸÑŸÖÿßÿ±ŸÉŸàŸÅŸäÿ©

This example demonstrates:
1. MDP structure
2. States, actions, and rewards
3. Transition probabilities
4. Policy representation
"""

print("=" * 60)
print("Example 1: Markov Decision Process (MDP)")
print("ŸÖÿ´ÿßŸÑ 1: ÿπŸÖŸÑŸäÿ© ÿßŸÑŸÇÿ±ÿßÿ± ÿßŸÑŸÖÿßÿ±ŸÉŸàŸÅŸäÿ©")
print("=" * 60)

# Simple MDP: Grid World
# MDP ÿ®ÿ≥Ÿäÿ∑: ÿπÿßŸÑŸÖ ÿßŸÑÿ¥ÿ®ŸÉÿ©
print("\n1. MDP Components")
print("ŸÖŸÉŸàŸÜÿßÿ™ MDP")
print("-" * 60)

# States: positions in a 3x3 grid
# ÿßŸÑÿ≠ÿßŸÑÿßÿ™: ÿßŸÑŸÖŸàÿßÿ∂ÿπ ŸÅŸä ÿ¥ÿ®ŸÉÿ© 3x3
states = {
    (0, 0): "Start",
    (0, 1): "State 1",
    (0, 2): "State 2",
    (1, 0): "State 3",
    (1, 1): "State 4",
    (1, 2): "State 5",
    (2, 0): "State 6",
    (2, 1): "State 7",
    (2, 2): "Goal"
}

# Actions: up, down, left, right
# ÿßŸÑÿ•ÿ¨ÿ±ÿßÿ°ÿßÿ™: ÿ£ÿπŸÑŸâÿå ÿ£ÿ≥ŸÅŸÑÿå Ÿäÿ≥ÿßÿ±ÿå ŸäŸÖŸäŸÜ
actions = ['up', 'down', 'left', 'right']

# Rewards: reward for each state
# ÿßŸÑŸÖŸÉÿßŸÅÿ¢ÿ™: ŸÖŸÉÿßŸÅÿ£ÿ© ŸÑŸÉŸÑ ÿ≠ÿßŸÑÿ©
rewards = {
    (0, 0): -0.1,
    (0, 1): -0.1,
    (0, 2): -0.1,
    (1, 0): -0.1,
    (1, 1): -0.1,
    (1, 2): -0.1,
    (2, 0): -0.1,
    (2, 1): -0.1,
    (2, 2): 10.0  # Goal state
}

print("\nStates:")
for state, description in states.items():
    reward = rewards[state]
    print(f"  {state}: {description} (reward: {reward})")

print(f"\nActions: {actions}")

# Transition probabilities (simplified)
# ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ÿßŸÑÿßŸÜÿ™ŸÇÿßŸÑ (ŸÖÿ®ÿ≥ÿ∑ÿ©)
def get_next_state(state, action):
    """
    Get next state after taking action (simplified).
    ÿßŸÑÿ≠ÿµŸàŸÑ ÿπŸÑŸâ ÿßŸÑÿ≠ÿßŸÑÿ© ÿßŸÑÿ™ÿßŸÑŸäÿ© ÿ®ÿπÿØ ÿßÿ™ÿÆÿßÿ∞ ÿßŸÑÿ•ÿ¨ÿ±ÿßÿ° (ŸÖÿ®ÿ≥ÿ∑).
    """
    row, col = state
    
    if action == 'up' and row > 0:
        return (row - 1, col)
    elif action == 'down' and row < 2:
        return (row + 1, col)
    elif action == 'left' and col > 0:
        return (row, col - 1)
    elif action == 'right' and col < 2:
        return (row, col + 1)
    
    return state  # Stay in same state if action invalid

# Example: Starting from (0,0) and moving right
print("\n" + "=" * 60)
print("2. Example Transition")
print("ŸÖÿ´ÿßŸÑ ÿπŸÑŸâ ÿßŸÑÿßŸÜÿ™ŸÇÿßŸÑ")
print("=" * 60)

start_state = (0, 0)
action_taken = 'right'
next_state = get_next_state(start_state, action_taken)

print(f"\nStart state: {start_state} ({states[start_state]})")
print(f"Action: {action_taken}")
print(f"Next state: {next_state} ({states[next_state]})")
print(f"Reward received: {rewards[next_state]}")

# Policy: mapping from states to actions
# ÿßŸÑÿ≥Ÿäÿßÿ≥ÿ©: ÿ™ÿπŸäŸäŸÜ ŸÖŸÜ ÿßŸÑÿ≠ÿßŸÑÿßÿ™ ÿ•ŸÑŸâ ÿßŸÑÿ•ÿ¨ÿ±ÿßÿ°ÿßÿ™
print("\n" + "=" * 60)
print("3. Policy Example")
print("ŸÖÿ´ÿßŸÑ ÿπŸÑŸâ ÿßŸÑÿ≥Ÿäÿßÿ≥ÿ©")
print("=" * 60)

# Simple policy: always move towards goal
# ÿ≥Ÿäÿßÿ≥ÿ© ÿ®ÿ≥Ÿäÿ∑ÿ©: ÿßŸÑÿ™ÿ≠ÿ±ŸÉ ÿØÿßÿ¶ŸÖÿßŸã ŸÜÿ≠Ÿà ÿßŸÑŸáÿØŸÅ
policy = {
    (0, 0): 'right',
    (0, 1): 'right',
    (0, 2): 'down',
    (1, 0): 'right',
    (1, 1): 'right',
    (1, 2): 'down',
    (2, 0): 'right',
    (2, 1): 'right',
    (2, 2): None  # Goal state, no action needed
}

print("\nPolicy (state -> action):")
for state, action in policy.items():
    if action:
        print(f"  {state}: {action}")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ÿ™ŸÖ ÿ•ŸÉŸÖÿßŸÑ ÿßŸÑŸÖÿ´ÿßŸÑ ÿ®ŸÜÿ¨ÿßÿ≠!")
print("=" * 60)

```
