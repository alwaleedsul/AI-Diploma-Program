## Step 2: Implementation | ÿßŸÑÿÆÿ∑Ÿàÿ© 2: ÿßŸÑÿ™ŸÜŸÅŸäÿ∞

**BEFORE**: We understand the concept.

**AFTER**: We'll see HOW it's implemented!


```python
# Example 1: Statistical Inference
# WHY: Need to make statistical conclusions
# HOW: Use hypothesis testing and confidence intervals

# Simulate model performance
np.random.seed(42)
train_scores = np.random.normal(0.85, 0.03, 100)  # Mean 85%, std 3%
val_scores = np.random.normal(0.83, 0.02, 100)     # Mean 83%, std 2%

# Compute means and confidence intervals
train_mean = np.mean(train_scores)
train_std = np.std(train_scores)
val_mean = np.mean(val_scores)
val_std = np.std(val_scores)

n = len(train_scores)
train_ci = 1.96 * train_std
np.sqrt(n)  # 95% CI
val_ci = 1.96 * val_std
np.sqrt(n)

print("Example 1: Statistical Inference")
print("=" * 60)
print(f"Training accuracy: {train_mean:.2%} ¬± {train_ci:.2%} (95% CI)")
print(f"Validation accuracy: {val_mean:.2%} ¬± {val_ci:.2%} (95% CI)")
print(f"\nüí° WHY: Quantify uncertainty in model performance")
print(f"üí° HOW: Use confidence intervals (mean ¬± 1.96√óSE)")
print(f"üí° AFTER: Can provide uncertainty estimates!")
```

---

## üìä Visualization | ÿßŸÑÿ™ÿµŸàÿ±


```python
# Visualize confidence intervals
fig, ax = plt.subplots(figsize=(10, 6))
means = [train_mean, val_mean]
cis = [train_ci, val_ci]
labels = ['Training', 'Validation']
colors = ['green', 'orange']

ax.barh(labels, means, xerr=cis, color=colors, alpha=0.7, capsize=10)
ax.set_xlabel('Accuracy', fontsize=12)
ax.set_title('Model Performance with 95% Confidence Intervals', fontsize=14, fontweight='bold')
ax.set_xlim(0.75, 0.90)
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print("\nüìä Visualization:")
print("  - Bars show mean accuracy")
print("  - Error bars show 95% confidence intervals")
```
