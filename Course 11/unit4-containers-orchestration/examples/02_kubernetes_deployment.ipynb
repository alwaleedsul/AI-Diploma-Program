## Part 2: Deployment YAML Example


```python
yaml_example = '''
apiVersion: apps/v1
kind: Deployment
metadata:
  name: ml-model-service
spec:
  replicas: 3
  selector:
    matchLabels:
      app: ml-model
  template:
    metadata:
      labels:
        app: ml-model
    spec:
      containers:
      - name: model-server
        image: my-registry/ml-model:latest
        ports:
        - containerPort: 8000
'''

print('ğŸ“ Deployment YAML:')
print(yaml_example)
print('\nâœ… Deployment configuration understood!')
```
