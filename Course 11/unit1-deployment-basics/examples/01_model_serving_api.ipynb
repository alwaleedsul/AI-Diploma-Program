## Code Example | Ù…Ø«Ø§Ù„ Ø§Ù„ÙƒÙˆØ¯

```python
"""
Unit 1 - Example 1: Model Serving with REST API
Ø§Ù„ÙˆØ­Ø¯Ø© 1 - Ù…Ø«Ø§Ù„ 1: ØªÙ‚Ø¯ÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… REST API

This example demonstrates:
1. Saving and loading a model
2. Creating a REST API endpoint
3. Serving predictions via API
"""

import joblib
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

print("=" * 60)
print("Example 1: Model Serving with REST API")
print("Ù…Ø«Ø§Ù„ 1: ØªÙ‚Ø¯ÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… REST API")
print("=" * 60)

# 1. Train and Save Model
# ØªØ¯Ø±ÙŠØ¨ ÙˆØ­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
print("\n1. Training and Saving Model")
print("ØªØ¯Ø±ÙŠØ¨ ÙˆØ­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬")
print("-" * 60)

# Create sample data
X, y = make_classification(n_samples=100, n_features=4, random_state=42)

# Train model
model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

print("âœ“ Model trained successfully")
print("âœ“ ØªÙ… ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ù†Ø¬Ø§Ø­")

# Save model
model_filename = 'trained_model.pkl'
joblib.dump(model, model_filename)
print(f"âœ“ Model saved to {model_filename}")
print(f"âœ“ ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ ÙÙŠ {model_filename}")

# Load model
loaded_model = joblib.load(model_filename)
print("âœ“ Model loaded successfully")
print("âœ“ ØªÙ… ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ù†Ø¬Ø§Ø­")

# 2. API Structure (using Flask)
# Ù‡ÙŠÙƒÙ„ API (Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… Flask)
print("\n" + "=" * 60)
print("2. REST API Structure")
print("Ù‡ÙŠÙƒÙ„ REST API")
print("=" * 60)

api_code = """
from flask import Flask, request, jsonify
import joblib
import numpy as np

app = Flask(__name__)

# Load model at startup
model = joblib.load('trained_model.pkl')

@app.route('/predict', methods=['POST'])
def predict():
    # Get data from request
    data = request.json
    features = np.array(data['features']).reshape(1, -1)
    
    # Make prediction
    prediction = model.predict(features)[0]
    probability = model.predict_proba(features)[0].tolist()
    
    return jsonify({
        'prediction': int(prediction), 'probability': probability
    })

@app.route('/health', methods=['GET'])
def health():
    return jsonify({'status': 'healthy'})

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000, debug=True)
"""

print(api_code)

# 3. Example API Request
# Ù…Ø«Ø§Ù„ Ø¹Ù„Ù‰ Ø·Ù„Ø¨ API
print("\n" + "=" * 60)
print("3. Example API Request")
print("Ù…Ø«Ø§Ù„ Ø¹Ù„Ù‰ Ø·Ù„Ø¨ API")
print("=" * 60)

example_request = """
POST http://localhost:5000/predict
Content-Type: application/json

{
    "features": [0.5, 0.3, 0.8, 0.2]
}
"""

print(example_request)

example_response = """
Response:
{
    "prediction": 1, "probability": [0.2, 0.8]
}
"""

print(example_response)

# Test the loaded model
print("\n" + "=" * 60)
print("4. Testing Loaded Model")
print("Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø§Ù„Ù…Ø­Ù…Ù‘Ù„")
print("=" * 60)

test_sample = np.array([[0.5, 0.3, 0.8, 0.2]])
prediction = loaded_model.predict(test_sample)
probability = loaded_model.predict_proba(test_sample)

print(f"\nTest sample: {test_sample[0]}")
print(f"Prediction: {prediction[0]}")
print(f"Probability: {probability[0]}")

print("\n" + "=" * 60)
print("Example completed successfully!")
print("ØªÙ… Ø¥ÙƒÙ…Ø§Ù„ Ø§Ù„Ù…Ø«Ø§Ù„ Ø¨Ù†Ø¬Ø§Ø­!")
print("=" * 60)
print("\nNote: Install Flask to run API:")
print("pip install flask")

```
